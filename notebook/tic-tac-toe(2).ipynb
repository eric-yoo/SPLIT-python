```python
msgs=["1번 플레이어의 차례입니다.","2번 플레이어의 차례입니다."]
ends=["1번 플레이어가 승리하였습니다.","2번 플레이어가 승리하였습니다."]
marks=['O','X']
board=[]
for i in range(3):
    board.append([])
    for j in range(3):
        board[i].append('_')
        
for i in range(3):
        for j in range(3):
            print(board[i][j], end='\t')
        print('')

# TODO: 위 코드를 적절히 가져다 써서 틱택토 게임을 만들어주세요~

def get_input(board): #함수
    try: #예외 처리
        user_input=int(input())
        if user_input in range(1,10):
            if board[2-(user_input-1)//3][(user_input-1)%3]=='_':
                return user_input # return = 함수의 output
            else: # 1~9 입력을 했는데 칸이 이미 차있는 경우
                print('칸이 이미 차 있습니다.')
                return get_input(board)
        else: # 숫자는 맞는데 1~9 입력이 아니다
            print('1~9 숫자가 아닙니다.')
            return get_input(board)
                
    except: # 형변환 에러 (숫자가 아니다)
        print('잘못된 입력입니다.')
        return get_input(board)
    
        
flag=False
for iter in range(9):
    mark=marks[iter%2]
    print(msgs[iter%2])

    user_input=get_input(board)
    
    x=2-(user_input-1)//3
    y=(user_input-1)%3
    board[x][y]=marks[iter%2]
    board_t=list(map(list, zip(*board)))

    for i in range(3):
        for j in range(3):
            print(board[i][j], end='\t')
        print('')
    
###     승리판단 코드
    for i in range(3): 
        if board[i]==[mark,mark,mark]: # 가로
            flag=True 
        if board_t[i]==[mark,mark,mark]: # 세로
            flag=True
    for i in range(3): # 대각선 \
        if board[i][i]!=mark:
            break
        if i==2:
            flag=True
    for i in range(3): # 대각선 /
        if board[2-i][i]!=mark:
            break
        if i==2:
            flag=True
    
    if flag:
        print(ends[iter%2])
        break
```
