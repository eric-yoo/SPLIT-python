```python
city_loc_list=city_location.values.tolist()
```

```python
sum=0
for label in city_loc_list:
    if label[0]=="SACRAMENTO": #도시명이 sacremanto면
        sum+=float(label[1]) #위도를 더해라
print(sum)

city_location_by_city.loc['SACRAMENTO'].std() #인덱스가 SACR~인 행들의 평균
```
